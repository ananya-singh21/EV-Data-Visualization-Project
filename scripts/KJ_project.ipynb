```python

import plotly.express as px
```

```python
dflong = dfcop.melt(id_vars="region_country", var_name="Country", value_name="ev Sales")
fig = px.choropleth(
    dflong,
    locations="Country",         # country names
    locationmode="country names",
    color="ev Sales",               # heat value
    hover_name="Country",
    animation_frame="region_country",      # animation by year
    color_continuous_scale="YlOrRd"
)

fig.update_layout(
    title="EV Sales by Country Over Years",
    geo=dict(showframe=False, showcoastlines=False)
)

fig.show()
```
